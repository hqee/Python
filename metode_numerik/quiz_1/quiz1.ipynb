**Quiz 1 : Dengan menggunakan bahasa Pyhton buatlah progrm untuk menyelesaiakan persamaan non Linear dari X3 + X2 - 3X- 3 pada selang [1,2] dengan 100 iterasi dan epsilon= 0.0004**

**1. Metode Tabel**

```python
def f(x):
    # Fungsi f(x) yang didefinisikan sesuai dengan persamaan yang diberikan.
    # f(x) = x^3 + x^2 - 3x - 3
    return x**3 + x**2 - 3*x - 3

# Rentang nilai x yang akan diperiksa
x_mulai = 1     # Nilai awal x
x_selesai = 2   # Nilai akhir x

iterations = 100   # Jumlah iterasi yang akan dilakukan
epsilon = 0.0004   # Toleransi kesalahan yang digunakan (tidak digunakan dalam metode tabel)

# Menghitung perubahan kecil (delta_x) untuk setiap iterasi berdasarkan rentang nilai x dan jumlah iterasi
delta_x = (x_selesai - x_mulai) / iterations

x_n = x_mulai   # Memulai perhitungan dengan x_n sebagai nilai awal

# Mencetak header untuk output tabel
print("Iterasi\tXn\t\tf(Xn)\t\tXn+1\t\tf(Xn+1)\t\tf(Xn)*f(Xn+1)")

# Inisialisasi variabel untuk menyimpan informasi jika akar ditemukan
root_found = False       # Menandai apakah akar sudah ditemukan
root_iteration = None    # Menyimpan iterasi saat akar ditemukan
root_x_n = None          # Menyimpan nilai x_n saat akar ditemukan
root_x_next = None       # Menyimpan nilai x_next saat akar ditemukan

# Melakukan iterasi sebanyak 'iterations' kali
for i in range(iterations):
    x_next = x_n + delta_x   # Menghitung x berikutnya (x_n+1)
    f_xn = f(x_n)            # Menghitung nilai fungsi di x_n
    f_xnext = f(x_next)      # Menghitung nilai fungsi di x_n+1
    
    # Mencetak hasil iterasi saat ini
    print(f"{i+1}\t{x_n:.6f}\t{f_xn:.6f}\t{x_next:.6f}\t{f_xnext:.6f}\t{(f_xn * f_xnext):.6f}")
    
    # Memeriksa perubahan tanda antara f(x_n) dan f(x_n+1)
    if not root_found and f_xn * f_xnext < 0:
        # Jika ada perubahan tanda, maka akar berada di antara x_n dan x_n+1
        root_found = True      # Tandai bahwa akar telah ditemukan
        root_iteration = i + 1 # Catat iterasi di mana akar ditemukan
        root_x_n = x_n         # Catat nilai x_n saat ini
        root_x_next = x_next   # Catat nilai x_next saat ini

    # Update x_n menjadi x_next untuk iterasi berikutnya
    x_n = x_next

# Memeriksa apakah akar ditemukan atau tidak
if root_found:
    # Jika akar ditemukan, cetak informasi tentang iterasi dan nilai-nilai x
    print("\nPerubahan tanda ditemukan pada iterasi", root_iteration)
    print(f"x_n = {root_x_n:.6f}, x_n+1 = {root_x_next:.6f}")
else:
    # Jika tidak ditemukan perubahan tanda setelah semua iterasi, cetak pesan ini
    print("\nTidak ditemukan perubahan tanda dalam 100 iterasi.")
```

**2. Metode Biseksi**

```python
def f(x):
    # Fungsi f(x) yang didefinisikan sesuai dengan persamaan yang diberikan.
    # f(x) = x^3 + x^2 - 3x - 3
    return x**3 + x**2 - 3*x - 3

# Rentang nilai awal untuk pencarian akar
a = 1  # Titik awal interval
b = 2  # Titik akhir interval

iterations = 100  # Jumlah iterasi maksimal yang akan dilakukan
epsilon = 0.0004  # Toleransi kesalahan yang digunakan untuk menentukan konvergensi

# Mencetak header untuk output tabel
print("Metode Biseksi")
print("Iterasi\t a\t\t b\t\t c\t\t f(a)\t\t f(b)\t\t f(c)")

# Inisialisasi variabel untuk menyimpan informasi jika akar ditemukan
root_found_bisection = False     # Menandai apakah akar sudah ditemukan
root_iteration_bisection = None  # Menyimpan iterasi saat akar ditemukan
root_c_bisection = None          # Menyimpan nilai c (akar perkiraan) saat akar ditemukan

# Melakukan iterasi sebanyak 'iterations' kali atau sampai akar ditemukan
for i in range(iterations):
    c = (a + b) / 2  # Menghitung titik tengah interval (c)
    fa = f(a)        # Menghitung nilai fungsi di titik a
    fb = f(b)        # Menghitung nilai fungsi di titik b
    fc = f(c)        # Menghitung nilai fungsi di titik tengah (c)
    
    # Mencetak hasil iterasi saat ini
    print(f"{i+1}\t{a:.6f}\t{b:.6f}\t{c:.6f}\t{fa:.6f}\t{fb:.6f}\t{fc:.6f}")
    
    # Memeriksa apakah nilai fungsi di titik tengah c mendekati nol
    if not root_found_bisection and abs(fc) < epsilon:
        # Jika nilai absolut f(c) kurang dari epsilon, akar ditemukan
        root_found_bisection = True       # Tandai bahwa akar telah ditemukan
        root_iteration_bisection = i + 1  # Catat iterasi di mana akar ditemukan
        root_c_bisection = c              # Catat nilai c (akar perkiraan)

    # Memperbarui interval untuk iterasi berikutnya berdasarkan tanda nilai fungsi
    if fa * fc < 0:
        # Jika f(a) dan f(c) memiliki tanda yang berbeda, akar berada di interval [a, c]
        b = c  # Memperbarui titik akhir interval menjadi c
    else:
        # Jika f(c) dan f(b) memiliki tanda yang berbeda, akar berada di interval [c, b]
        a = c  # Memperbarui titik awal interval menjadi c

# Memeriksa apakah akar ditemukan atau tidak
if root_found_bisection:
    # Jika akar ditemukan, cetak informasi tentang iterasi dan nilai c
    print(f"\nAkar ditemukan dengan metode biseksi pada iterasi {root_iteration_bisection} dengan c = {root_c_bisection:.6f}")
else:
    # Jika tidak ditemukan akar setelah semua iterasi, cetak pesan ini
    print("\nTidak ditemukan akar dalam 100 iterasi dengan metode biseksi.")
```

**3. Metode Regular Falsi**

```python
def f(x):
    # Fungsi f(x) yang didefinisikan sesuai dengan persamaan yang diberikan.
    # f(x) = x^3 + x^2 - 3x - 3
    return x**3 + x**2 - 3*x - 3

# Rentang nilai awal untuk pencarian akar
a = 1  # Titik awal interval
b = 2  # Titik akhir interval

iterations = 100  # Jumlah iterasi maksimal yang akan dilakukan
epsilon = 0.0004  # Toleransi kesalahan yang digunakan untuk menentukan konvergensi

# Mencetak header untuk output tabel
print("Metode Regula Falsi")
print("Iterasi\t a\t\t b\t\t c\t\t f(a)\t\t f(b)\t\t f(c)")

# Inisialisasi variabel untuk menyimpan informasi jika akar ditemukan
root_found_regula_falsi = False     # Menandai apakah akar sudah ditemukan
root_iteration_regula_falsi = None  # Menyimpan iterasi saat akar ditemukan
root_c_regula_falsi = None          # Menyimpan nilai c (akar perkiraan) saat akar ditemukan

# Melakukan iterasi sebanyak 'iterations' kali atau sampai akar ditemukan
for i in range(iterations):
    # Menghitung titik c dengan menggunakan interpolasi linier
    fa = f(a)  # Menghitung nilai fungsi di titik a
    fb = f(b)  # Menghitung nilai fungsi di titik b
    c = b - (fb * (a - b)) / (fa - fb)  # Titik baru berdasarkan interpolasi linier
    fc = f(c)  # Menghitung nilai fungsi di titik c
    
    # Mencetak hasil iterasi saat ini
    print(f"{i+1}\t{a:.6f}\t{b:.6f}\t{c:.6f}\t{fa:.6f}\t{fb:.6f}\t{fc:.6f}")
    
    # Memeriksa apakah nilai fungsi di titik c mendekati nol
    if not root_found_regula_falsi and abs(fc) < epsilon:
        # Jika nilai absolut f(c) kurang dari epsilon, akar ditemukan
        root_found_regula_falsi = True       # Tandai bahwa akar telah ditemukan
        root_iteration_regula_falsi = i + 1  # Catat iterasi di mana akar ditemukan
        root_c_regula_falsi = c              # Catat nilai c (akar perkiraan)

    # Memperbarui interval untuk iterasi berikutnya berdasarkan tanda nilai fungsi
    if fa * fc < 0:
        # Jika f(a) dan f(c) memiliki tanda yang berbeda, akar berada di interval [a, c]
        b = c  # Memperbarui titik akhir interval menjadi c
    else:
        # Jika f(c) dan f(b) memiliki tanda yang berbeda, akar berada di interval [c, b]
        a = c  # Memperbarui titik awal interval menjadi c

# Memeriksa apakah akar ditemukan atau tidak
if root_found_regula_falsi:
    # Jika akar ditemukan, cetak informasi tentang iterasi dan nilai c
    print(f"\nAkar ditemukan dengan metode regula falsi pada iterasi {root_iteration_regula_falsi} dengan c = {root_c_regula_falsi:.6f}")
else:
    # Jika tidak ditemukan akar setelah semua iterasi, cetak pesan ini
    print("\nTidak ditemukan akar dalam 100 iterasi dengan metode regula falsi.")
```

**4. Metode Secant**

```python
# Definisikan fungsi
def f(x):
    return x**3 + x**2 - 3*x - 3

# Parameter
x0 = 1  # Nilai awal x0
x1 = 2  # Nilai awal x1
n = 100  # Jumlah iterasi maksimum
epsilon = 0.0004  # Toleransi kesalahan

# List untuk menyimpan hasil setiap iterasi
iterations = []
x0_values = []
x1_values = []
x2_values = []
f_x2_values = []

# Metode Secant
for i in range(1, n + 1):
    # Hitung x2 menggunakan metode Secant
    x2 = x1 - (f(x1) * (x1 - x0)) / (f(x1) - f(x0))
    f_x2 = f(x2)
    
    # Mencatat hasil setiap iterasi
    iterations.append(i)
    x0_values.append(x0)
    x1_values.append(x1)
    x2_values.append(x2)
    f_x2_values.append(f_x2)
    
    # Pengecekan akar atau penghentian jika error kurang dari epsilon
    if abs(f_x2) < epsilon:
        root_found = True
        root_x2 = x2
        root_f_x2 = f_x2
        break
    else:
        root_found = False
    
    # Memperbarui nilai x0 dan x1 untuk iterasi berikutnya
    x0 = x1
    x1 = x2

# Menampilkan iterasi
print(f"{'Iterasi':<10} {'x0':<15} {'x1':<15} {'x2':<15} {'f(x2)':<15}")
for i in range(len(iterations)):
    print(f"{iterations[i]:<10} {x0_values[i]:<15.6f} {x1_values[i]:<15.6f} {x2_values[i]:<15.6f} {f_x2_values[i]:<15.6f}")

# Menampilkan hasil akhir
if root_found:
    print(f"\nAkar ditemukan di x = {root_x2:.6f} dengan f(x) = {root_f_x2:.6f}")
else:
    print("\nAkar tidak ditemukan dalam iterasi yang diberikan.")
```
